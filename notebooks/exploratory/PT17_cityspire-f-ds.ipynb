```python
# Import pandas and set display options.

import pandas as pd
pd.options.display.max_rows = 999
```

```python
city_list_df = pd.read_csv('/content/cities_list_df.csv')
print(city_list_df.shape)
city_list_df.head()
```

```python
# Create a set of items from Hailey's city list.

hailey_set = set(city_list_df['Cities_List'])
print(len(hailey_set))
```

```python
# Load up partially cleaned up dataframe from the previous cohort. 
# This one had cities and states cleaned up from the metro areas.

labor_df = pd.read_csv('https://raw.githubusercontent.com/Lambda-School-Labs/PT17_cityspire-f-ds/main/notebooks/datasets/data/labor_statistics/split_labor.csv')
print(labor_df.shape)
labor_df.head()
```

```python
# Combine the City and State columns into a new column.
# State column had a leading space, which was not cleaned.

labor_df["Cities_list"] = labor_df['city'] + "," + labor_df['state']
```

```python
# Print the length of the strings that match Akron OH in both dataframes
# in order to make sure that they match in length.

print(len(city_list_df['Cities_List'].iloc[0]))
print(len(labor_df['Cities_list'].iloc[3]))
```

```python
# Create a set that matches Hailey's city list format.

adi_set = set(labor_df['Cities_list'])
print(len(adi_set))
```

```python
# Create an intersection set. 

intersection_set = hailey_set.intersection(adi_set)
print(len(intersection_set))
```

```python
# Convert it into a list so we can sort it.

labor_city_list = list(intersection_set)
```

```python
# Sort the list. 

labor_city_list.sort()
```

```python
# Place it into a dataframe.

labor_cities_df = pd.DataFrame(labor_city_list, columns=['Cities_List'])
labor_cities_df.head()
```

```python
# Save it as csv file.

labor_cities_df.to_csv(path_or_buf='/content/result.csv', index=False)
```
