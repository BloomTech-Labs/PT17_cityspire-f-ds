```python
import pandas as pd
import numpy as np
import os
import os.path
import regex
from datetime import datetime


# Loading original data set 

FILE_PATH = os.path.abspath(os.path.dirname('data/final.csv'))
cities = pd.read_csv(FILE_PATH + '/final.csv')

#Replacing the names of cities with underscores 
cities.replace('\s+', '_',regex=True,inplace=True)
cities
```

```python
# Load data set retrieved from visual crossing first batch 
FILE_PATH = os.path.abspath(os.path.dirname('data/combined.csv'))
cities_df = pd.read_csv(FILE_PATH + '/combined.csv')
cities_df = cities_df.drop(columns=['Name'])
cities_df.head()
```

```python
# Convert strings to Date time
cities_df['Date time'] = pd.to_datetime(cities_df['Date time'], errors='coerce',format ='%d/%m/%Y')

cities_df.head()
```

```python
# Load data set retrieved from visual crossing second batch 
FILE_PATH = os.path.abspath(os.path.dirname('data/weather_data_merged.csv'))
cities_df1 = pd.read_csv(FILE_PATH + '/weather_data_merged.csv')
cities_df1 = cities_df1.drop(columns=['Name.1'])
cities_df1.head()
```

```python
# Renamed column to match the other dataset and to be able to merge them on a primary key 
cities_df1 = cities_df1.rename(columns={"Name": "Location"})
cities_df1.head()
```

```python
# Convert strings to Date time
cities_df1['Date time'] = pd.to_datetime(cities_df1['Date time'], errors='coerce',format ='%d/%m/%Y') 
cities_df1.tail()
```

```python
cities_df.columns
```

```python
# Renaming columns to match the first dataset 
cities_df1.columns
cities_df1 = cities_df1.rename(columns={'Cloud Cover':'Cloud Cover (%)', 'Minimum Temperature': 'Minimum Temperature (degF)', 
                          'Precipitation': 'Precipitation (in)', 'Solar Radiation': 'Solar Radiation (W/m^2)',
                          'Dew Point': 'Dew Point (degF)', 'Relative Humidity': 'Relative Humidity (%)', 
                          'Precipitation Cover': 'Precipitation Cover (%)','Temperature': 'Temperature (degF)',
                           'Maximum Temperature': 'Maximum Temperature (degF)', 'Visibility': 'Visibility (mi)',
                           'Wind Speed': 'Wind Speed (mph)', 'Solar Energy': 'Solar Energy (J/m^2)', 
                           'Heat Index': 'Heat Index (degF)', 'Snow Depth': 'Snow Depth (in)', 
                           'Sea Level Pressure': 'Sea Level Pressure (mb)', 'Snow': 'Snow (in)', 
                           'Wind Gust': 'Wind Gust (mph)','Wind Chill': 'Wind Chill (degF)'})

cities_df1.head(3)
```

```python
cities_df.shape, cities_df1.shape
```

```python
# Merging 2 datasets on 'city', which is a primary key. 
# cities_combined = pd.merge(cities_df, cities_df1, on='Location', how='outer')
# cities_combined

combined_df = pd.concat([cities_df, cities_df1], sort=True)
combined_df.shape
```

```python
# Split Location column into 3: City, State and Country. Since all the cities are from the USA, we do not need Country
combined_df[['City','State','Country']] = combined_df['Location'].str.split(", ", expand=True)
combined_df.head()
```

```python
# Check if there are any null/NaN values
combined_df.isnull().sum()
```

```python
# Rename columns. Include only relevant ones and remove the ones categorical variables that contain a lot of missing value
combined_df = combined_df[['City', 'State', 'Date time', 'Latitude','Longitude','Temperature (degF)',
                           'Minimum Temperature (degF)','Maximum Temperature (degF)','Relative Humidity (%)',
                           'Cloud Cover (%)','Precipitation (in)','Precipitation Cover (%)', 'Wind Direction', 
                           'Wind Speed (mph)', 'Wind Gust (mph)', 'Wind Chill (degF)','Snow Depth (in)', 'Snow (in)',
                           'Solar Radiation (W/m^2)', 'Dew Point (degF)','Visibility (mi)','Solar Energy (J/m^2)', 
                           'Heat Index (degF)', 'Sea Level Pressure (mb)']]
combined_df.head(3)
```

```python
combined_df.dtypes
```

```python
# Function to convert DF columns to floats
def convert_to_float(combined_df):
    combined_df = combined_df.astype(float)
    return combined_df
```

```python
# Apply convert_to_float function to convert strings into floats  
combined_df['Latitude'] = convert_to_float(combined_df['Latitude'])
combined_df['Longitude'] = convert_to_float(combined_df['Longitude'])
combined_df['Temperature (degF)'] = convert_to_float(combined_df['Temperature (degF)'])
combined_df['Minimum Temperature (degF)'] = convert_to_float(combined_df['Minimum Temperature (degF)'])
combined_df['Maximum Temperature (degF)'] = convert_to_float(combined_df['Maximum Temperature (degF)'])
combined_df['Relative Humidity (%)'] = convert_to_float(combined_df['Relative Humidity (%)'])
combined_df['Cloud Cover (%)'] = convert_to_float(combined_df['Cloud Cover (%)'])
combined_df['Precipitation (in)'] = convert_to_float(combined_df['Precipitation (in)'])
combined_df['Precipitation Cover (%)'] = convert_to_float(combined_df['Precipitation Cover (%)'])
combined_df['Wind Direction'] = convert_to_float(combined_df['Wind Direction'])
combined_df['Wind Speed (mph)'] = convert_to_float(combined_df['Wind Speed (mph)'])
combined_df['Wind Gust (mph)'] = convert_to_float(combined_df['Wind Gust (mph)'])
combined_df['Wind Chill (degF)'] = convert_to_float(combined_df['Wind Chill (degF)'])
combined_df['Snow Depth (in)'] = convert_to_float(combined_df['Snow Depth (in)'])
combined_df['Solar Radiation (W/m^2)'] = convert_to_float(combined_df['Solar Radiation (W/m^2)'])
combined_df['Dew Point (degF)'] = convert_to_float(combined_df['Dew Point (degF)'])
combined_df['Snow (in)'] = convert_to_float(combined_df['Snow (in)'])
combined_df['Solar Radiation (W/m^2)'] = convert_to_float(combined_df['Solar Radiation (W/m^2)'])
combined_df['Dew Point (degF)'] = convert_to_float(combined_df['Dew Point (degF)'])
combined_df['Visibility (mi)'] = convert_to_float(combined_df['Visibility (mi)'])
combined_df['Solar Energy (J/m^2)'] = convert_to_float(combined_df['Solar Energy (J/m^2)'])
combined_df['Heat Index (degF)'] = convert_to_float(combined_df['Heat Index (degF)'])
combined_df['Sea Level Pressure (mb)'] = convert_to_float(combined_df['Sea Level Pressure (mb)'])
```
