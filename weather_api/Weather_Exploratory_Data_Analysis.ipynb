```python
"""astype float was giving me errors, so I have tp comvert them one by one using to_numeric 
pandas function because it only accepted either Series or 1d array"""

combined_df['Latitude'] = pd.to_numeric(combined_df['Latitude'], errors='coerce')
combined_df['Longitude'] = pd.to_numeric(combined_df['Longitude'], errors='coerce')
combined_df['Temperature (degF)'] = pd.to_numeric(combined_df['Temperature (degF)'], errors='coerce')
combined_df['Minimum Temperature (degF)'] = pd.to_numeric(combined_df['Minimum Temperature (degF)'], errors='coerce')
combined_df['Maximum Temperature (degF)'] = pd.to_numeric(combined_df['Maximum Temperature (degF)'], errors='coerce')
combined_df['Relative Humidity (%)'] = pd.to_numeric(combined_df['Relative Humidity (%)'], errors='coerce')
combined_df['Cloud Cover (%)'] = pd.to_numeric(combined_df['Cloud Cover (%)'], errors='coerce')
combined_df['Precipitation (in)'] = pd.to_numeric(combined_df['Precipitation (in)'], errors='coerce')
combined_df['Precipitation Cover (%)'] = pd.to_numeric(combined_df['Precipitation Cover (%)'], errors='coerce')
combined_df['Wind Direction'] = pd.to_numeric(combined_df['Wind Direction'], errors='coerce')
combined_df['Wind Speed (mph)'] = pd.to_numeric(combined_df['Wind Speed (mph)'], errors='coerce')
combined_df['Wind Gust (mph)'] = pd.to_numeric(combined_df['Wind Gust (mph)'], errors='coerce')
combined_df['Wind Chill (degF)'] = pd.to_numeric(combined_df['Wind Chill (degF)'], errors='coerce')
combined_df['Snow Depth (in)'] = pd.to_numeric(combined_df['Snow Depth (in)'], errors='coerce')
combined_df['Solar Radiation (W/m^2)'] = pd.to_numeric(combined_df['Solar Radiation (W/m^2)'], errors='coerce')
combined_df['Dew Point (degF)'] = pd.to_numeric(combined_df['Dew Point (degF)'], errors='coerce')
combined_df['Snow (in)'] = pd.to_numeric(combined_df['Snow (in)'], errors='coerce')
combined_df['Solar Radiation (W/m^2)'] = pd.to_numeric(combined_df['Solar Radiation (W/m^2)'], errors='coerce')
combined_df['Dew Point (degF)'] = pd.to_numeric(combined_df['Dew Point (degF)'], errors='coerce')
combined_df['Visibility (mi)'] = pd.to_numeric(combined_df['Visibility (mi)'], errors='coerce')
combined_df['Solar Energy (J/m^2)'] = pd.to_numeric(combined_df['Solar Energy (J/m^2)'], errors='coerce')
combined_df['Heat Index (degF)'] = pd.to_numeric(combined_df['Heat Index (degF)'], errors='coerce')
combined_df['Sea Level Pressure (mb)'] = pd.to_numeric(combined_df['Sea Level Pressure (mb)'], errors='coerce')
```
