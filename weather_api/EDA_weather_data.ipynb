```python
from wwo_hist import retrieve_hist_data
import pandas as pd
import os
import os.path
import regex

# Loading original data set 

FILE_PATH = os.path.abspath(os.path.dirname('other_data/final.csv'))
cities = pd.read_csv(FILE_PATH + '/final.csv')
cities.replace('\s+', '_',regex=True,inplace=True)
cities
```

```python
# Load data set retrievd from weatheronline
FILE_PATH = os.path.abspath(os.path.dirname('merged_cities.csv'))
cities_df = pd.read_csv(FILE_PATH + '/merged_cities.csv')
cities_df
```

```python
# Renamed column to match the other dataset and to be able to merge them on a primary key 
cities_df = cities_df.rename(columns={"location": "City"})
cities_df
```

```python
# Merging 2 datasets on 'city', which is a primary key. 
cities_combined = pd.merge(cities, cities_df, on='City', how='outer')
cities_combined
```
